### Assessing Summary
The dataset contains 569 rows or observation and 32 columns,
There are some columns with missing values, 
5 Rows in dataset is duplicated.

### Cleaning
We only need 12 columns from the 32 column, 
We will handle and drop dulpicates.

```python
new_df=cancer_data.iloc[:,:12]
new_df.head()
```

```python
new_df.columns[2][:-5]
```

```python
new_col=[]
for col in new_df.columns:
    if '_mean' in col:
        new_col.append(col[:-5])
    else:
        new_col.append(col)

new_col
```

```python
new_df.columns=new_col
```

```python
new_df.head()
```

```python
## Handle the duplicates values
new_df.drop_duplicates(inplace=True)
```

```python
new_df.duplicated().sum()
```

```python
new_df.isnull().sum()
```

```python
##Handling missing values
new_df['smoothness'].fillna(new_df['smoothness'].mean(),inplace=True)
```

```python
new_df['symmetry'].fillna(new_df['symmetry'].mean(),inplace=True)
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.isnull().sum()
```

```python
## save
new_df.to_csv('cancer_cleaned.csv', index=False)
```
